```python
import os
import sys


directory_path = os.path.abspath('../')

if directory_path not in sys.path:
    sys.path.append(directory_path)

from analyze import get_spectral_phase_colors

colors = get_spectral_phase_colors(ratemaps, grid_scores, mask_threshold=0.0)
```

```python
from matplotlib import pyplot as plt
import numpy as np


side = int(np.sqrt(len(colors)))
topo_map = colors.reshape(side, side, 3)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(topo_map, interpolation='nearest')
plt.title("Topographic Phase Map for Non-Topographic Model")
plt.axis('off')
plt.show()
```
